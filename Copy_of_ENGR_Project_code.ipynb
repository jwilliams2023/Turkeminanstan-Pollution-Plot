```python
# Vandan Patadia
# Elawoe Dumoga
# Graph 1
import matplotlib.pyplot as plt

import numpy as np

# Data
x = [1990, 1995, 2000, 2005, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]

y = [24.69, 27.12, 28.73, 21.94, 26.10, 32.28, 30.77, 30.56, 30.62, 28.62, 27.51, 27.13, 26.55, 26.20]

# Izayah tubigon
# setting up the plot
plt.figure(figsize=(12, 6))

plt.plot(x, y, marker='o', linestyle='-')

plt.xticks(x, rotation=45)

# labels and titles and font

plt.title('Mean Annual Pollution', fontsize=20)

plt.xlabel('Year', fontsize=14)

plt.ylabel('Air Pollution (micrograms per cubic meter)', fontsize=12)

# save the plot and displays
plt.savefig('Turkmenistan')
```
